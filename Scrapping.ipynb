```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import seaborn as sns
%matplotlib inline
import re
import time
from datetime import datetime
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = 'https://www.zameen.com/Homes/Lahore-1-2.html'
hrml=requests.get(url);
soup = BeautifulSoup(hrml.text, 'lxml')
```

```python
print(soup.prettify());
```

```python
pricex = soup.find('span', {'class': 'f343d9ce', 'aria-label':'Listing price'})
print(pricex.text);
    
```

```python
namex = soup.find('div', {'class': '_162e6469', 'aria-label':'Listing location'})
print(namex.text);
```

```python
for detail in soup.find_all('div', {'class': '_162e6469', 'aria-label':'Listing location'}):
    print(detail.text);
```

```python
location=[];
for cp in np.arange(1, 300):
    url = 'https://www.zameen.com/Homes/Karachi-2-'+  str(cp) +'.html'
    hrml=requests.get(url);
    soup = BeautifulSoup(hrml.text, 'lxml')
    
    for detail in soup.find_all('div', {'class': '_162e6469', 'aria-label':'Listing location'}):
        print(detail.text);
        location.append(detail.text);
```

```python
cdata = {"Location":location}
df = pd.DataFrame(data = cdata)
df.to_csv("C:\\Users\\Fozia\\Desktop\\Data.csv",index=False, header=True)
```
